```python
date_col= pd.DatetimeIndex(df["DT"])
df["Year"]= date_col.year
df["Month"] =date_col.month
df["Day"] = date_col.day
df
```

```python
date_col= pd.DatetimeIndex(df["time"])
df["Hour"]= date_col.hour
df["Minutes"] =date_col.minute
df["Second"] = date_col.second
df
```

```python
df[["road", "street","city", "zip"]] = df["ADDR"].str.split(",", expand=True)
df["road"] = df["road"].str.strip()
df["street"] = df["street"].str.strip()
df["city"]= df["city"].str.strip()
df["zip"] = df["zip"].str.strip()
df
```

```python
#Deleting column from df
df = df.drop('ADDR', axis=1)
df
```

```python
df = df.drop('DT', axis=1)
df
```
